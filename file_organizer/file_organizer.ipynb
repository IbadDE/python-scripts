```python

# Import required libraries. 
import os
import shutil


#Create folder if it doesn't exists
def create_folder(folder_name, destination_location):
    folder_path = os.path.join(destination_location, folder_name)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        return os.path.join(destination_location, folder_name)
    else:
        return os.path.join(destination_location, folder_name)
        


def move_files(source_location, destination_location):
    pdf_folder = create_folder('PDF Files', destination_location)
    images_folder = create_folder('Images', destination_location)
    datasets_folder = create_folder('Datasets', destination_location)
    zip_folder = create_folder('Zip file', destination_location)
    docs_folder = create_folder('word file', destination_location)
    tableau_folder = create_folder('Tableau', destination_location)
    Others = create_folder('Other', destination_location)

    for root, dirs, files in os.walk(source_location):
        for file in files:
            source_file_path = os.path.join(root, file)
            if file.lower().endswith('.pdf'):                                # moving pdf file
                destination_pdf = os.path.join(pdf_folder, file)
                try:
                    shutil.move(source_file_path, destination_pdf)
                except Exception as e:
                    print(f"file {file} did't moved: {e}") 
            elif file.lower().endswith(('.jpg', '.png')):                     #for images
                destination_images = os.path.join(images_folder, file)
                try:
                    shutil.move(source_file_path, destination_images)
                except Exception as e:
                    print(f"file {file} did't moved: {e}")
            elif file.lower().endswith(('.xlsx', '.csv', '.xls')):               # for excels files
                destination_datasets = os.path.join(datasets_folder, file)
                try:
                    shutil.move(source_file_path, destination_datasets)
                except Exception as e:
                    print(f"file {file} did't moved: {e}") 
            elif file.lower().endswith('.zip'):                                  # for zip files
                destination_zip = os.path.join(zip_folder, file)
                try:
                    shutil.move(source_file_path, destination_zip)
                except Exception as e:
                    print(f"file {file} did't moved: {e}") 
            elif file.lower().endswith(('.docx','.txt')):                           # for word and text 
                destination_word = os.path.join(docs_folder, file)
                try:
                    shutil.move(source_file_path, destination_word)
                except Exception as e:
                    print(f"file {file} did't moved: {e}") 
            elif file.lower().endswith(('.twbx')):                                   # for tableau files
                destination_tableau = os.path.join(tableau_folder, file)
                try:
                    shutil.move(source_file_path, destination_tableau)
                except Exception as e:
                    print(f"file {file} did't moved: {e}") 
            else :                                                                   # rest of the files
                destination_others = os.path.join(Others, file)
                try:
                    shutil.move(source_file_path, destination_others)
                except Exception as e:
                    print(f"file {file} did't moved: {e}") 
                
    

```

```python
move_files(r'C:\Users\Ibad\Downloads', r'C:\Users\Ibad\Desktop')
```
